# Khaiii 사용한 전처리 연습

## Import Packages 

```python
import re # 레귤러 익스프레션??
# import urllib.request
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

## Import Data

```python
# 뉴스 데이터 불러오기
test_table = pd.read_excel('/content/drive/MyDrive/yg_test.xlsx').drop('Unnamed: 0', axis=1)
test_table
```

## 데이터 정제

### 한글만 남기기

```python
# 데이터에서 한글 제외하고 모두 제거
# 공란 제거
test_table['art_title'] = test_table['art_title'].str.replace("[^ㄱ-ㅎㅏ-ㅣ가-힣 ]", " ")
test_table['art_main'] = test_table['art_main'].str.replace("[^ㄱ-ㅎㅏ-ㅣ가-힣 ]", " ")
test_table['art_title'].str.strip().replace("", np.nan,)
test_table['art_main'].str.strip().replace("", np.nan)
test_table = test_table.dropna(how='any')
test_table.to_csv('yg_test')
```

### 형태소 분석 + 품사 선택

```python
# 형태소 분류하고 원하는 품사만 선택하여 태깅된 품사 리스트 만들기
from khaiii import KhaiiiApi
api = KhaiiiApi()

# 주요 품사
significant_tags = ['NNG', 'NNP', 'NNB', 'VV', 'VA', 'VX', 'MAG', 'MAJ', 'XSV', 'XSA']


X_test = []
for s in test_table['art_main']:                # 표에서 기사
    temp=[]
    for w in api.analyze(s):                    # 기사에서 단어
        for morph in w.morphs:                  # 형태소 표현
            if morph.tag in significant_tags:   # 주요 품사만 선택
                temp.append(morph.lex + '/' + morph.tag)
    X_test.append(temp)
```

```python
print(X_test[0])
```

### 동사를 원형으로 치환

```python
# 동사 원형으로 돌려주는 기능을 가진 함수
def stem_form(articles):
    p1 = re.compile('[가-힣A-Za-z0-9]+/NN. [가-힣A-Za-z0-9]+/XS.')
    p2 = re.compile('[가-힣A-Za-z0-9]+/NN. [가-힣A-Za-z0-9]+/XSA [가-힣A-Za-z0-9]+/VX')
    p3 = re.compile('[가-힣A-Za-z0-9]+/VV')
    p4 = re.compile('[가-힣A-Za-z0-9]+/VX')
    corpus = []
    for sent in articles:
        ori_sent = sent
        mached_terms = re.findall(p1, ori_sent)
        for terms in mached_terms:
            ori_terms = terms
            modi_terms = ''
            for term in terms.split(' '):
                lemma = term.split('/')[0]
                tag = term.split('/')[-1]
                modi_terms += lemma
            modi_terms += '다/VV'
            ori_sent = ori_sent.replace(ori_terms, modi_terms)
        
        mached_terms = re.findall(p2, ori_sent)
        for terms in mached_terms:
            ori_terms = terms
            modi_terms = ''
            for term in terms.split(' '):
                lemma = term.split('/')[0]
                tag = term.split('/')[-1]
                if tag != 'VX':
                    modi_terms += lemma
            modi_terms += '다/VV'
            ori_sent = ori_sent.replace(ori_terms, modi_terms)

        mached_terms = re.findall(p3, ori_sent)
        for terms in mached_terms:
            ori_terms = terms
            modi_terms = ''
            for term in terms.split(' '):
                lemma = term.split('/')[0]
                tag = term.split('/')[-1]
                modi_terms += lemma
            if '다' != modi_terms[-1]:
                modi_terms += '다'
            modi_terms += '/VV'
            ori_sent = ori_sent.replace(ori_terms, modi_terms)

        mached_terms = re.findall(p4, ori_sent)
        for terms in mached_terms:
            ori_terms = terms
            modi_terms = ''
            for term in terms.split(' '):
                lemma = term.split('/')[0]
                tag = term.split('/')[-1]
                modi_terms += lemma
            if '다' != modi_terms[-1]:
                modi_terms += '다'
            modi_terms += '/VV'
            ori_sent = ori_sent.replace(ori_terms, modi_terms)
        corpus.append(ori_sent)
    return corpus

def to_stem(articles):
    for i in range(len(articles)):
        articles[i] = stem_form(articles[i])
    return articles
```

```python
# 동사 원형으로 돌림
X_test = to_stem(X_test)
print(X_test[0])
```

### 불용어 처리

```python
def remove_stopword_text(articles, stopwords):
    corpus = []
    for sent in articles:
        modi_sent = []
        for word in sent.split(' '):
            if word not in stopwords:
                modi_sent.append(word)
        corpus.append(' '.join(modi_sent))
    return corpus

def remove_stopword(articles, stopwords):
    for i in range(len(articles)):
        articles[i] = remove_stopword_text(articles[i], stopwords)
    return articles
```

```python
stopwords = ['데/NNB', '좀/MAG', '수/NNB', '등/NNB']
X_test = remove_stopword(X_test, stopwords)
print(X_test[0])
```

### 토큰

```python
def tokentoken(articles):
    tokenizer = Tokenizer()
    tokenizer.fit_on_texts(articles)
    # 빈도수 낮은 단어수 체크
    threshold = 3
    words_cnt = len(tokenizer.word_index)
    rare_cnt = 0
    words_freq = 0
    rare_freq = 0

    for key, value in tokenizer.word_counts.items():
        words_freq = words_freq + value
        if value < threshold:
            rare_cnt += 1
            rare_freq = rare_freq + value
    print("전체단어수", words_cnt)
    print(f"빈도수 {threshold-1} 이하인 단어수:", rare_cnt)
    print("희귀 단어 비율:", rare_cnt / words_cnt )
    print("희귀 단어 등장 빈도 비율:", rare_freq / words_freq)
    # 제거
    vocab_size = words_cnt - rare_cnt + 2
    tokenizer = Tokenizer(vocab_size, oov_token='OOV')
    tokenizer.fit_on_texts(articles)
    # 토큰화
    articles = tokenizer.texts_to_sequences(articles)
    return articles
```

```python
X_test = tokentoken(X_test)
```
