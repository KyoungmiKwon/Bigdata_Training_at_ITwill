    2) 크롤링 한 뉴스 리스트 모두 스크랩 후 저장 하기

```python
news_article={}
idx = 1

for i, (date, info, title) in enumerate(zip(dates, infos, titles)):    
    
    date = date.text.strip()
    info = info.text.strip()
    url_article = 'https://finance.naver.com' + title['href']
    title = title.text.strip()

    res_a = requests.get(url_article)
    html_a = res_a.text.strip()
    soup_a = BeautifulSoup(html_a, features='html5lib')

    select_head = 'strong.media_end_summary' # 헤드라인만
    select_news = 'div#news_read' # 헤드라인 + 본문 (잡소리포함)

    head = soup_a.select(select_head)
    news = soup_a.select(select_news)
        
    for line in news:
        body = line.text.strip()

        # dit 형태로 파일 변환
        news_article[idx]={'date': date, 'info': info, 'title': title, 'body': body}        

        print(f'[{idx}]',date,info,title,'----본문----',body)
        idx += 1

# DataFrame으로 변환
news_df = pd.DataFrame(news_article)
# excel 파일로 저장 https://blog.naver.com/xowns55/222052797454
news_df.to_excel('C:\Lab\kyoungmi_test.xlsx')
```

# 여러 페이지 스크랩 하기

```python
#> '122870' : YG
#> '035900' : JYP
#> '005930' : 삼성전자
#> '066570' : LG전자

code = ['122870'] #> 기입할것!
pages = 20        #> 기입할것!


url = 'https://finance.naver.com/item/news_news.nhn?&sm=title_entity_id.basic&clusterId=' # URL

news_article={}
idx = 1

for i in range(pages+1):
    params ={'code': code, 'page':i}
    response = requests.get(url,params)
    html = response.text.strip()
    soup = BeautifulSoup(markup=html, features='html5lib')

    # CSS route 확인
    select_date = 'tr > td.date'
    select_from = 'tr > td.info'
    select_title = 'tr > td > a.tit'

    # 해당 관련 부분 선택
    dates = soup.select(select_date)
    infos = soup.select(select_from)
    titles = soup.select(select_title)

    for j, (date, info, title) in enumerate(zip(dates, infos, titles)):
        date = date.text.strip()
        info = info.text.strip()
        url_article = 'https://finance.naver.com' + title['href']
        title = title.text.strip()

        res_a = requests.get(url_article)
        html_a = res_a.text.strip()
        soup_a = BeautifulSoup(html_a, features='html5lib')

        select_head = 'strong.media_end_summary' # 헤드라인만
        select_news = 'div#news_read' # 헤드라인 + 본문 (잡소리포함분)

        head = soup_a.select(select_head)
        news = soup_a.select(select_news)
            
        for line in news:
            body = line.text.strip()

            # dit 형태로 파일 변환
            news_article[idx]={'date': date, 'info': info, 'title': title, 'body': body}
            
            print(f'[{idx}]',date,info,title,'----본문----',body)
            idx += 1
    
# DataFrame으로 변환
news_df = pd.DataFrame(news_article)

# excel 파일로 저장
news_df.to_excel('C:\Lab\kyoungmi_test1.xlsx')
```
